```python
import face_recognition
import cv2
```

```python
# input video file

input_movie = cv2.VideoCapture("demo.mp4")
length = int(input_movie.get(cv2.CAP_PROP_FRAME_COUNT))
print(length)

```

```python
# output video file

fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_movie = cv2.VideoWriter('result.avi', fourcc, 29.97, (1280, 720))
```

```python
# input image
image_1 = face_recognition.load_image_file("Mahershala Ali.jpg")
first_face = face_recognition.face_encodings(image_1)[0]

image_2 = face_recognition.load_image_file("Sam Elliott.jpg")
second_face = face_recognition.face_encodings(image_2)[0]

image_3 = face_recognition.load_image_file("Richard E Grant.jpg")
third_face = face_recognition.face_encodings(image_3)[0]

image_4 = face_recognition.load_image_file("Adam driver.jpg")
forth_face = face_recognition.face_encodings(image_4)[0]

image_5 = face_recognition.load_image_file("Sam_Rockwell.jpg")
fifth_face = face_recognition.face_encodings(image_5)[0]

image_6 = face_recognition.load_image_file("Joanne Tucker.jpg")
sixth_face = face_recognition.face_encodings(image_6)[0]

image_7= face_recognition.load_image_file("Leslie Bibb.jpg")
seventh_face = face_recognition.face_encodings(image_7)[0]

image_8= face_recognition.load_image_file("Amatus Sami-Karim.jpg")
eighth_face = face_recognition.face_encodings(image_8)[0]

image_9= face_recognition.load_image_file("Viggo Mortensen.jpg")
nineth_face = face_recognition.face_encodings(image_9)[0]

image_10= face_recognition.load_image_file("Octavia Spencer.jpg")
tenth_face = face_recognition.face_encodings(image_10)[0]
```

```python
# known_faces

known_faces = [
    first_face,
    second_face,
    third_face,
    forth_face,
    fifth_face,
    sixth_face,
    seventh_face,
    eighth_face,
    nineth_face,
    tenth_face
    
   
]
```

```python
font = cv2.FONT_HERSHEY_DUPLEX
face_locations = []
face_encodings = []
face_names = []
frame_number = 0
red = (0, 0, 255)
white = (255, 255, 255)

while True:

    ret, frame = input_movie.read()
    frame_number += 1

    # exit if input file ends
    if not ret:
        break

    # Convert BGR (OpenCV) to RGB (face_recognition)
    rgb_frame = frame[:, :, ::-1]

    # face encodings in the current frame of video
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    face_names = []
    for face_encoding in face_encodings:
        # find a match
        match = face_recognition.compare_faces(known_faces, face_encoding, tolerance=0.50)

        name = None
        if match[0]:
            name = "Mahershala Ali"
        elif match[1]:
            name = "Sam Elliott"
        elif match[2]:
            name = "Richard E Grant"
        elif match[3]:
            name = "Adam driver"
        elif match[4]:
            name = "Sam_Rockwell"
        elif match[5]:
            name = "Joanne Tucker"
        elif match[6]:
            name = "Leslie Bibb"
        elif match[7]:
            name = "Amatus Sami-Karim"
        elif match[8]:
            name = "Viggo Mortensen"
        elif match[9]:
            name = "Octavia Spencer"
        face_names.append(name)
        
            # write the name
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        color = red
    	
        if not name:
            name = "Unknown"
            color = white

        # draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), color, 1)

        # put a name below the face
        cv2.putText(frame, name, (left, bottom + 15), font, 0.5, color, 1)

    # write the result to the output file
    output_movie.write(frame)
    if(frame_number % 100 == 0):
    	print("Frame {} of {} - {}%".format(frame_number, length, '%.2f' % (frame_number/length*1)))
    

print('Result video file is ready.')
input_movie.release()
```
